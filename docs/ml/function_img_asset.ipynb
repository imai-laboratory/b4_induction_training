```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 1, 0.01)
y = 1 / (1 - x)

plt.title("odds(p) = 1 / (1 - p)")
plt.xlabel("p")
plt.ylabel("odds(p)")
plt.plot(x, y)
# plt.show()
plt.savefig("odds.png")
```

```python
x = np.arange(0, 1, 0.01)
y = np.log( x / (1 - x) )

plt.title("logit(p) = log ( p / (1 - p) )")
plt.xlabel("p")
plt.ylabel("logit(p)")
plt.plot(x, y)
# plt.show()
plt.savefig("logit.png")
```

```python
x = np.arange(-5, 5, 0.01)
y = 1/(1 + np.exp(-x))

plt.title("Sigmoid(x) =  1 / (1 + exp(-x))")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x, y)
# plt.show()
plt.savefig("sigmoid.png")
```
